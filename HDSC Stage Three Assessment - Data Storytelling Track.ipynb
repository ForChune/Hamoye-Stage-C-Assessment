```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
#import food production data and transform it into a dataset
fprod_df = pd.read_csv('./Hamoye/Africa Food Production (2004 - 2013).csv')
```

```python
#filter rice produced in African countries
rice_prod_df = fprod_df.loc[(fprod_df['Item'] == 'Rice (Milled Equivalent)')]
rice_prod_df.head()
```

```python
#Question 1 - Which country produced the highest amount of oats in 2010? 
oats_prod_df = fprod_df.loc[(fprod_df['Item'] == 'Oats')]
oats_prod_df_2010 = oats_prod_df[oats_prod_df['Year']==2010]
oats_prod_df_2010.head()
```

__The answer is Algeria__

```python
#Question 2 - Which three countries respectively had the lowest food supply between 2004 and 2006 (inclusive)?
#import food supply dataset
fdsupply_df = pd.read_csv('./Hamoye/Africa Food Supply (2004 - 2013).csv')

sup_line_plot = sns.lineplot(x = fdsupply_df['Year'], y = fdsupply_df['Value'], hue = fdsupply_df['Country'], palette = 'bright', ci = None, data = fdsupply_df)
# sup_line_plot.set_yscale('log')
sup_line_plot.legend(loc='upper right', bbox_to_anchor=(1.6, 1.1))
sup_line_plot.set(Title='Food Supply in Africa')
```

__The visualization shows that Zambia, Ethiopia and Central African Republic had the lowest food supply between 2004 and 2006__

```python
#Question 5 - What three countries have produced the highest total quantity of rice from 2004 to 2013?
#filter rice produced in African countries
rice_prod_df = fprod_df.loc[(fprod_df['Item'] == 'Rice (Milled Equivalent)')]
rice_df = rice_prod_df.groupby(['Country'])['Value'].sum()
rice_df.sort_values(ascending=False)
```
